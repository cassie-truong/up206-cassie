```python
# Dataset description
# This dataset contains locations and boundaries for properties that Philadelphia Parks and Recreation (PPR) has resposibility for or has a distinct role in the maintenance or management.
```

```python
# importing geopandas instead of pandas because the pandas function can't read shapefiles
import geopandas as gpd
```

```python
# adding shapefile to notebook
PPR = gpd.read_file('data/PPR_Properties.zip')
```

```python
# analyzing the columns available in this data
PPR.columns.to_list()
```

```python
# generating values for a single column (public name of park)
PPR.PUBLIC_NAM.sample(5)
```

```python
# generating values for a single column (zip code)
PPR.ZIP_CODE.sample(5)
```

```python
# analyzing the count of parks ine ach zip code
PPR['ZIP_CODE'].value_counts()
```

```python
# saving the zip code count as a variable
zip_count = PPR['ZIP_CODE'].value_counts()
zip_count
```

```python
# testing to see if the variable saved
zip_count
```

```python
# creating a bar graph showing the number of parks per zip code
zip_count.plot.bar(x = 'zip code', y = 'count', title = 'Number of parks per zip code')
```

```python
#plotting a map of the parks in all zip codes
PPR_trimmed.plot(
    figsize=(20,12),
    column = 'ZIP_CODE',
    legend = True,
    legend_kwds={
        'loc':'lower right',
        'bbox_to_anchor':(1,1)
    }
)
```
